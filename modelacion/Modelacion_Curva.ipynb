```python
import numpy as np 
import matplotlib.pyplot as plt 
from funciones.curvaturas import get_radius
```

```python
# Inicial
x_0 = 300
y_0 = 1100

# Inflexión
x1 = 1000
y1 = 1400

# Crítico
x2 = 1400
y2 = 2200

# Final
x_f = 1800
y_f = 1200
```

```python
x = np.array([x_0,x1,x2,x_f])
y = np.array([y_0,y1,y2,y_f])
z = np.polyfit(x,y,3)
p = np.poly1d(z)
print (np.poly1d(p))
z
```

```python
def ilustracion(x,y,z,p):
    xp= np.linspace(x_0,x_f+100,100)
    plt.ylim(0,y_f+1500)
    plt.plot(x, y, '.',xp,p(xp),'-')
    plt.style.use("ggplot")

    return plt.show()

ilustracion(x,y,z,p)
```

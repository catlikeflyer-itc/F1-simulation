# Sumas de Riemann
Asumiendo la funci√≥n como:
$f(x) = x^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : x**2 # Funci√≥n Lambda para la ecuaci√≥n
a = 0 # L√≠mite inferior
b = 5 # L√≠mite superior
delta = 0.2 # Delta

x = np.arange(a,b+delta,delta)
y = f(x)
print(sum(y*delta))

# Se utilizan para graficar
X = np.arange(a,b+delta,delta)
Y = f(X)

plt.figure(figsize=(15,5))

# Tomando el punto izquierdo
plt.subplot(1,3,1)
plt.plot(X,Y,'b')
x_left = x[:-1] 
y_left = y[:-1]
plt.plot(x_left,y_left,'b.',markersize=10)
plt.bar(x_left,y_left,width=(b-a)/len(x),alpha=0.2,align='edge',edgecolor='b')
plt.title('Suma de Riemann del punto izquierdo')

# Tomando el punto medio
plt.subplot(1,3,2)
plt.plot(X,Y,'b')
x_mid = (x[:-1] + x[1:])/2
y_mid = f(x_mid)
plt.plot(x_mid,y_mid,'b.',markersize=10)
plt.bar(x_mid,y_mid,width=(b-a)/len(x),alpha=0.2,edgecolor='b')
plt.title('Suma de Riemann del punto medio')

# Tomando el punto derecho
plt.subplot(1,3,3)
plt.plot(X,Y,'b')
x_right = x[1:] 
y_right = y[1:]
plt.plot(x_right,y_right,'b.',markersize=10)
plt.bar(x_right,y_right,width=-(b-a)/len(x),alpha=0.2,align='edge',edgecolor='b')
plt.title('Suma de Riemann del punto derecho')

plt.show()
```
